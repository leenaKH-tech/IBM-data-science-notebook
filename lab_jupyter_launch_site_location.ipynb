Next, let's create markers for all launch records.
If a launch was successful `(class=1)`, then we use a green marker and if a launch was failed, we use a red marker `(class=0)`


Note that a launch only happens in one of the four launch sites, which means many launch records will have the exact same coordinate. Marker clusters can be a good way to simplify a map containing many markers having the same coordinate.


Let's first create a `MarkerCluster` object


```python
marker_cluster = MarkerCluster()
```

*TODO:* Create a new column in `spacex_df` dataframe called `marker_color` to store the marker colors based on the `class` value


```python

# Apply a function to check the value of `class` column
# If class=1, marker_color value will be green
# If class=0, marker_color value will be red
# Answering the geographical questions:
print("1. All launch sites are near the Equator (between 28°-34°N latitude) - optimal for rocket launches due to Earth's rotation boost.")
print("2. All sites are coastal (Florida and California) - safer for failed launches and lower population risk.")

# Create marker color column in spacex_df
spacex_df['marker_color'] = spacex_df['class'].apply(lambda x: 'green' if x == 1 else 'red')

# Create MarkerCluster for launch records
marker_cluster = MarkerCluster().add_to(site_map)

# Add colored markers to cluster
for index, record in spacex_df.iterrows():
    folium.Marker(
        location=[record['Lat'], record['Long']],
        icon=folium.Icon(color=record['marker_color']),
        popup=f"Site: {record['Launch Site']}<br>Outcome: {'Success' if record['class'] == 1 else 'Failure'}"
    ).add_to(marker_cluster)

# Display enhanced map
site_map
```

Next, we need to explore and analyze the proximities of launch sites.


Let's first add a `MousePosition` on the map to get coordinate for a mouse over a point on the map. As such, while you are exploring the map, you can easily find the coordinates of any points of interests (such as railway)


```python
# Add Mouse Position to get the coordinate (Lat, Long) for a mouse over on the map
formatter = "function(num) {return L.Util.formatNum(num, 5);};"
mouse_position = MousePosition(
    position='topright',
    separator=' Long: ',
    empty_string='NaN',
    lng_first=False,
    num_digits=20,
    prefix='Lat:',
    lat_formatter=formatter,
    lng_formatter=formatter,
)

site_map.add_child(mouse_position)
site_map
```

```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

*TODO:* Mark down a point on the closest coastline using MousePosition and calculate the distance between the coastline point and the launch site.


```python
# find coordinate of the closet coastline
# e.g.,: Lat: 28.56367  Lon: -80.57163
# distance_coastline = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)
# Assume spacex_df has columns: ['Lat', 'Long', 'Launch Site', ...]
coastline_coords = {
    "CCAFS LC-40": (28.56398, -80.56809),  # Example coastline near Cape Canaveral
    "VAFB SLC-4E": (34.6329, -120.6155),   # Example coastline near Vandenberg
    # Add more if needed
}

# Compute distances for each launch site
spacex_df['Distance to Coastline (km)'] = spacex_df.apply(
    lambda row: calculate_distance(
        row['Lat'], 
        row['Long'], 
        *coastline_coords.get(row['Launch Site'], (0, 0))  # Default (0,0) if not found
    ),
    axis=1
)

print(spacex_df[['Launch Site', 'Distance to Coastline (km)']])
```

```python
# Create and add a folium.Marker on your selected closest coastline point on the map
# Display the distance between coastline point and launch site using the icon property 
# for example
# distance_marker = folium.Marker(
#    coordinate,
#    icon=DivIcon(
#        icon_size=(20,20),
#        icon_anchor=(0,0),
#        html='<div style="font-size: 12; color:#d35400;"><b>%s</b></div>' % "{:10.2f} KM".format(distance),
#        )
#    )
import folium
from folium.features import DivIcon

# Example: Coordinates for a launch site and its nearest coastline
launch_site_lat, launch_site_lon = 28.56367, -80.57163  # Cape Canaveral
coastline_lat, coastline_lon = 28.56398, -80.56809      # Nearest coastline point

# Calculate distance
distance_km = calculate_distance(launch_site_lat, launch_site_lon, coastline_lat, coastline_lon)

# Create a map centered at the launch site
site_map = folium.Map(location=[launch_site_lat, launch_site_lon], zoom_start=12)

# Add a marker for the launch site
folium.Marker(
    [launch_site_lat, launch_site_lon],
    popup="Launch Site",
    icon=folium.Icon(color='red', icon='rocket', prefix='fa')
).add_to(site_map)

# Add a marker for the coastline with distance displayed
distance_marker = folium.Marker(
    [coastline_lat, coastline_lon],
    icon=DivIcon(
        icon_size=(150, 30),  # Adjust size to fit text
        icon_anchor=(0, 0),
        html=f'<div style="font-size: 12pt; color: blue;"><b>{distance_km:.2f} KM</b></div>'
    )
)
distance_marker.add_to(site_map)

# Optionally, draw a line between the launch site and coastline
folium.PolyLine(
    locations=[[launch_site_lat, launch_site_lon], [coastline_lat, coastline_lon]],
    color='blue',
    weight=1.5,
    dash_array='5,5'
).add_to(site_map)

# Display the map
site_map
```

```python
# Create a `folium.PolyLine` object using the coastline coordinates and launch site coordinate
# lines=folium.PolyLine(locations=coordinates, weight=1)
#site_map.add_child(lines)
import folium
from folium.features import DivIcon

# 1. Define Coordinates (Example: Cape Canaveral)
launch_site = [28.56367, -80.57163]  # Launch Site Lat/Lon
coastline = [28.56398, -80.56809]    # Nearest Coastline Point

# 2. Calculate Distance
distance_km = calculate_distance(*launch_site, *coastline)

# 3. Create Map Centered at Launch Site
site_map = folium.Map(location=launch_site, zoom_start=12)

# 4. Add Launch Site Marker (Rocket Icon)
folium.Marker(
    launch_site,
    popup="<b>Launch Site</b>",
    icon=folium.Icon(color='red', icon='rocket', prefix='fa')
).add_to(site_map)

# 5. Add Coastline Distance Marker (Blue Text)
folium.Marker(
    coastline,
    icon=DivIcon(
        icon_size=(150, 30),
        icon_anchor=(0, 0),
        html=f'<div style="font-size: 12pt; color: blue;"><b>{distance_km:.2f} KM to Coast</b></div>'
    )
).add_to(site_map)

# 6. Draw Connecting Line (Dashed Blue)
lines = folium.PolyLine(
    locations=[launch_site, coastline],
    color='blue',
    weight=1.5,
    dash_array='5, 5'
)
site_map.add_child(lines)

# Display Map
site_map
```

```python
# Create a marker with distance to a closest city, railway, highway, etc.
# Draw a line between the marker to the launch site
import folium
from folium.features import DivIcon

# 1. Define Coordinates (Example: Cape Canaveral)
launch_site = [28.56367, -80.57163]  # Launch Site

# 2. Nearby Locations (with real approximate coordinates)
nearby_locations = {
    "City (Titusville)": [28.6122, -80.8076],
    "Highway (FL-401)": [28.5625, -80.5707],
    "Railway (Florida East Coast)": [28.5618, -80.5683]
}

# 3. Create Map
site_map = folium.Map(location=launch_site, zoom_start=12)

# 4. Add Launch Site Marker
folium.Marker(
    launch_site,
    popup="<b>SpaceX Launch Site</b>",
    icon=folium.Icon(color='red', icon='rocket', prefix='fa')
).add_to(site_map)

# 5. Process each nearby location
colors = ['green', 'blue', 'purple']  # Different colors for each type
for i, (location_name, coords) in enumerate(nearby_locations.items()):
    distance = calculate_distance(*launch_site, *coords)
    
    # Add location marker
    folium.Marker(
        coords,
        popup=f"<b>{location_name}</b>",
        icon=folium.Icon(color=colors[i], icon='info-sign')
    ).add_to(site_map)
    
    # Add distance label
    folium.Marker(
        coords,
        icon=DivIcon(
            icon_size=(200, 36),
            icon_anchor=(0, 0),
            html=f'<div style="font-size: 10pt; color: {colors[i]};">'
                 f'<b>{distance:.2f} km to {location_name}</b></div>'
        )
    ).add_to(site_map)
    
    # Draw connecting line
    folium.PolyLine(
        locations=[launch_site, coords],
        color=colors[i],
        weight=1.5,
        dash_array='5, 5'
    ).add_to(site_map)

# Display Map
site_map
```
